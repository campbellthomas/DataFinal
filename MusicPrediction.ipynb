```python
import re
def extract_average_weight(weight_str):
    if isinstance(weight_str, str):
        # Use regex to find all numbers in the string
        numbers = re.findall(r'\d+', weight_str)
        if len(numbers) == 2:
            num1 = int(numbers[0])
            num2 = int(numbers[1])
            
            return (num1 + num2) / 2
    return None  # Return None for non-string values or invalid formats

# Apply the function to the 'weight' column
df['average_weight'] = df['weight'].apply(extract_average_weight)

# Create a box plot to visualize the average weight by breed group
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['breed_group'] == breed]['average_weight'].dropna() for breed in df['breed_group'].unique()],
            labels=df['breed_group'].unique())
plt.title('Average Weight by Breed Group')
plt.xlabel('Breed Group')
plt.ylabel('Average Weight (pounds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

for breed in df['breed_group'].unique():
    print(breed)
    print(df[df['breed_group'] == breed]['average_weight'].describe())
```

```python
# Function to convert height strings to average height in inches

heights=df['height']
def extract_inches_range(height):
    matches = re.findall(r'\d+', height)
    inches = [int(match) for match in matches if int(match) <= 31]  
    return min(inches), max(inches)
inches_ranges = [extract_inches_range(height) for height in heights]

def find_avg(x,arr):
    for i in x:
        if i[0]==None:
            arr.append(None)
        else:
            arr.append(((i[1]+i[0])/2))

height_avg=[]
find_avg(inches_ranges,height_avg)
df['height_avg'] = height_avg


for breed in df['breed_group'].unique():
    print(breed)
    print(df[df['breed_group'] == breed]['height_avg'].describe())

# Create a box plot 
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['breed_group'] == breed]['height_avg'].dropna() for breed in df['breed_group'].unique()],
            labels=df['breed_group'].unique())
plt.title('Average Height by Breed Group')
plt.xlabel('Breed Group')
plt.ylabel('Average Height (inches)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python

breed_groups = df['breed_group'].unique()
variables = ['incredibly_kifriendly_dogs', 'dog_friendly', 'friendly_towarstrangers', 'all_around_friendliness']
names = ['Kid Friendly', 'Dog Friendly', 'Nice to Strangers', 'Overall Friendliness']

fig, axes = plt.subplots(2, 2, figsize=(12, 9))  # Create a 1x3 grid of plots

# Iterate over each variable to create a heatmap
for i, var in enumerate(variables):
    # Create a contingency table (cross-tabulation)
    contingency_table = pd.crosstab(df['breed_group'], df[var])
    
    # Plot heatmap
    row, col = divmod(i, 2)  # Determine row and column in the 2x2 grid
    cax = axes[row, col].matshow(contingency_table, cmap='Blues', aspect='auto')
    
    # Add colorbar
    fig.colorbar(cax, ax=axes[row, col])
    
     # Set labels and formatting
    axes[row, col].set_title(f'{names[i]} vs Breed Group', fontsize=20)
    axes[row, col].set_xticks(range(len(contingency_table.columns)))
    axes[row, col].set_yticks(range(len(breed_groups)))
    axes[row, col].set_xticklabels(contingency_table.columns, rotation=90, fontsize=14)
    axes[row, col].set_yticklabels(breed_groups, fontsize=14)
    axes[row, col].set_xlabel(names[i], fontsize=16)
    axes[row, col].set_ylabel('Breed Group', fontsize=16)

plt.tight_layout()
plt.savefig('heatmaps.png', transparent=True)
plt.show()
```

```python
#Preprocessing
#mode imputation 

# Group by 'breed_group' and calculate the mode for 'exercise_needs' in each group
mode_exercise_needs_by_breed = df.groupby('breed_group')['exercise_needs'].apply(lambda x: x.mode()[0] if not x.mode().empty else None)

# Fill missing values in 'exercise_needs' with the respective breed's mode
df['exercise_needs'] = df.apply(
    lambda row: mode_exercise_needs_by_breed[row['breed_group']] if pd.isnull(row['exercise_needs']) else row['exercise_needs'], axis=1
)

# Load your data - with missing because I forgot to check and already filled in 
df2 = pd.read_csv('data/data/training_data.csv')

# Check how many missing values in each breed
missing_exercise_by_breed = df2[df2['exercise_needs'].isnull()].groupby('breed_group')['exercise_needs'].size()

# Print breeds and how many missing values per breed
print("Breeds with missing 'exercise_needs' and the number of missing values:\n", missing_exercise_by_breed)

# Get the distribution of 'exercise_needs' before filling the missing values
plt.figure(figsize=(10,5))
plt.hist(df['exercise_needs'].dropna(), bins=range(1, 6), alpha=0.7, label='Before filling', color='blue', edgecolor='black')
plt.xlabel('Exercise Needs')
plt.ylabel('Frequency')
plt.title('Distribution of Exercise Needs Before Filling Missing Data')
plt.xticks(range(1, 6))  # Assuming the exercise needs range from 1 to 5
plt.show()


# Get the distribution of 'exercise_needs' after filling the missing values
plt.figure(figsize=(10,5))
plt.hist(df['exercise_needs'], bins=range(1, 6), alpha=0.7, label='After filling', color='green', edgecolor='black')
plt.xlabel('Exercise Needs')
plt.ylabel('Frequency')
plt.title('Distribution of Exercise Needs After Filling Missing Data')
plt.xticks(range(1, 6))  # Assuming the exercise needs range from 1 to 5
plt.show()

```
